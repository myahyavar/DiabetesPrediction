```python
# function to remove outliers
def clearOutliers(df,feature):
    upper_limit = df[feature].mean() + 3*df[feature].std() # Find the upper bound
    lower_limit = df[feature].mean() - 3*df[feature].std() #Find the lower bound
    #df[(df['BMI'] > upper_limit) | (df['BMI'] < lower_limit)]
    return df[(df[feature] < upper_limit) & (df[feature] > lower_limit)]
```

- In the first step, utilizing the default parameters of the RandomForest classifier yielded promising outcomes. By implementing optimizations and ensuring we steer clear of overfitting, we can attain exceptional results with this model.

- **In the next step, we have an imbalanced dataset where outliers are primarily associated with patients diagnosed with diabetes. To address this issue without removing outliers and avoiding overfitting, we can consider utilizing methods such as oversampling, undersampling, and bootstrapping. These techniques help balance the dataset by either replicating minority class instances (oversampling), randomly removing majority class instances (undersampling), or creating additional samples through resampling (bootstrapping). Additionally, we can explore other methods through research to further improve the model's performance. By incorporating these approaches, we aim to create a more effective model that handles the imbalanced dataset appropriately.Additionally, we will implement a Naive Bayes model from scratch**

```python
def RandomForestWithDepth(X,Y, X_test, y_test,max_depth):
    clf = RandomForestClassifier(max_depth=max_depth)
    x_scaled=sc.fit_transform(X)
    clf.fit(x_scaled, Y)
    X_test=sc.fit_transform(X_test)
    y_pred = clf.predict(X_test)
    return f1_score(y_test, y_pred, zero_division=1)
```

### The implementation of Naive Bayes model from scratch

```python
class MyGaussianNB:
    def __init__(self):
        self.p_0=0.0
        self.p_1=0.0
        self.p_no=0.0
        self.p_yes=0.0
        self.model_target = ""
        self.model_features = ""
        self.Gaussian_params = {}
    def fit(self, X_train, y_train):
        self.model_target = pd.DataFrame(data =y_train,columns=["target"])
        self.target_sum = len(self.model_target["target"])
        self.target_0 = len(self.model_target[self.model_target["target"] == 0])
        self.target_1 = len(self.model_target[self.model_target["target"] == 1])
        self.p_0 = self.target_0/self.target_sum # p(0) = sum(0)/sum(0,1)
        self.p_1 = self.target_1/self.target_sum # p(1) = sum(1)/sum(0,1)
        self.model_features = pd.DataFrame(data =X_train,columns=features)
        
        
        for feature in self.model_features:
            target_yes = self.model_features[(self.model_target["target"]==1)]#p(feature|target=1)
            #calculate the mean for this feature where the target is 1
            target_yes_mean = target_yes[feature].mean()
            #calculate the std for this feature where the target is 1
            target_yes_std = target_yes[feature].std()

            target_no  = self.model_features[(self.model_target["target"]==0)]#p(feature|target=0) 
            #calculate the mean for this feature where the target is 0
            target_no_mean = target_no[feature].mean()
            #calculate the std for this feature where the target is 0
            target_no_std = target_no[feature].std()
            
            self.Gaussian_params[feature] = {"mean_yes":target_yes_mean,"std_yes":target_yes_std,
                                             "mean_no":target_no_mean,"std_no":target_no_std}
            
    def predictOne(self, patient):
        self.p_no  = self.p_0
        self.p_yes = self.p_1
        for feature in self.model_features:
            params = self.Gaussian_params[feature]
            self.p_yes *= self.GaussianDistribution(patient[feature][0],params["std_yes"],params["mean_yes"])#p(target=1|feature)
            self.p_no *= self.GaussianDistribution(patient[feature][0],params["std_no"],params["mean_no"])#p(target=0|feature)
        if self.p_yes > self.p_no:
            return 1
        else:
            return 0
    def predict(self, X_test): 
        result_arr = []
        for row in X_test:
            patient = pd.DataFrame(data =[row],columns=features)
            result_arr.append(self.predictOne(patient))
        return result_arr
        
    def GaussianDistribution(self,x,std,mean):
        return 1/( std * np.sqrt(2 * np.pi)) * np.exp( - (x - mean)**2 / (2 * (std**2)) )
    
```

```python
def MyNaiveBayes(X_train, X_test, y_train, y_test):
    classifier = MyGaussianNB()
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    print("ConfusionMatrix:")
    print(confusion_matrix(y_test, y_pred))
    print("accuracy: ",accuracy_score(y_test, y_pred))
    print("f1_score: ",f1_score(y_test, y_pred, zero_division=1))
    return classifier
```

### Naive Bayes assumes that all features are independent, but in reality, it is highly unlikely to find a set of features that are completely independent. This limitation of Naive Bayes can result in suboptimal performance. On the other hand, Random Forest tends to provide better results because it employs decision trees in its algorithm.
